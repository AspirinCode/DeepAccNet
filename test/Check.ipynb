```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from os import listdir

digitizations = [-20.0, -15.0, -10.0, -4.0, -2.0, -1.0, -0.5, 0.5, 1.0, 2.0, 4.0, 10.0, 15.0, 20.0]
masses = [digitizations[0]]+[(digitizations[i]+digitizations[i+1])/2 for i in range(len(digitizations)-1)]+[digitizations[-1]]
```

```python
files = listdir("outputs1/")
file = files[0]
```

```python
temp = np.load("outputs1/"+file)
plt.figure(figsize=(8,1.5))
plt.plot(range(len(temp["lddt"])), temp["lddt"], label="true")
plt.show()

plt.figure(figsize=(6,3))
plt.subplot(121)
sns.heatmap(np.sum(temp["estogram"].transpose(1,2,0)*masses, axis=-1), cbar=False, cmap="seismic", vmin=-20, vmax=20)
plt.subplot(122)
sns.heatmap(temp["mask"], cbar=False, cmap="seismic", vmin=-1, vmax=1)
plt.show()
np.mean(temp["lddt"])
```
